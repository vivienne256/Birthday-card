```python
import os, random
from datetime import date
from PIL import Image
import streamlit as st

# ----------------------------
# Settings you should customize
# ----------------------------
LOVER_NAME = "Lily"   # <-- change
YOUR_NAME = "Yuanyuan"  # <-- change
BIRTHDAY = date(2026, 3, 5)  # <-- change (YYYY, M, D)

TITLE = f"Happy Birthday, {LOVER_NAME}! üéÇ"
OPENING = (
    "Today is all about you.\n\n"
    "Thank you for being my favourite person, my safe place, and my best surprise."
)

LOVE_NOTES = [
    "I love the way you laugh ‚Äî it instantly makes my day better.",
    "With you, even ordinary days feel special.",
    "You make me feel understood, supported, and loved.",
    "I‚Äôm proud of you ‚Äî always.",
]

PROMISES = [
    "More hugs. More photos. More sunsets together.",
    "More patience, more listening, more growing together.",
    "More dates, more adventures, more tiny everyday happiness.",
]

# Folder of photos
PHOTO_DIR = "photos"


# ----------------------------
# App UI
# ----------------------------
st.set_page_config(page_title=TITLE, page_icon="üéâ", layout="wide")

st.title(TITLE)
st.caption(f"Made with ‚ù§Ô∏è by {YOUR_NAME}")

# Countdown / status
today = date.today()
days = (BIRTHDAY - today).days
if days > 0:
    st.info(f"üéà {days} days until your birthday!")
elif days == 0:
    st.success("üéâ It's your birthday today!!!")
else:
    st.warning(f"üéÄ Happy belated birthday! (It was {-days} days ago)")

st.write(OPENING)

st.divider()

# Photo gallery
st.subheader("Our Memories üì∏")

photo_paths = []
if os.path.isdir(PHOTO_DIR):
    for fn in os.listdir(PHOTO_DIR):
        if fn.lower().endswith((".jpg", ".jpeg", ".png", ".webp")):
            photo_paths.append(os.path.join(PHOTO_DIR, fn))

photo_paths.sort()

if not photo_paths:
    st.warning("No photos found. Please put images into the 'photos/' folder.")
else:
    # Slider to pick photo
    idx = st.slider("Pick a photo", 1, len(photo_paths), 1) - 1
    img = Image.open(photo_paths[idx])
    st.image(img, use_container_width=True)

    # Optional: show thumbnails in columns
    with st.expander("Show all photos"):
        cols = st.columns(3)
        for i, p in enumerate(photo_paths):
            with cols[i % 3]:
                st.image(Image.open(p), use_container_width=True, caption=os.path.basename(p))

st.divider()

# Love notes section
st.subheader("A few things I love about you üíå")
st.write("Click for a random note:")

if "note" not in st.session_state:
    st.session_state.note = random.choice(LOVE_NOTES)

if st.button("Give me a love note"):
    st.session_state.note = random.choice(LOVE_NOTES)

st.success(st.session_state.note)

st.divider()

# Promises / wishes
st.subheader("My wish for you üåü")
st.write("May you be happy, healthy, and surrounded by love ‚Äî today and always.")
st.write("And here‚Äôs my promise:")

if "promise" not in st.session_state:
    st.session_state.promise = random.choice(PROMISES)

if st.button("Show my promise"):
    st.session_state.promise = random.choice(PROMISES)

st.info(st.session_state.promise)

st.divider()

# Hidden "easter egg"
st.subheader("One last thing‚Ä¶ ü´∂")
with st.expander("Open when you miss me"):
    st.write(
        f"Hey {LOVER_NAME},\n\n"
        "If you‚Äôre reading this, I want you to know:\n"
        "I‚Äôm always on your side.\n\n"
        "Come here ‚Äî you deserve a big hug. ü§ç"
    )

st.caption("Tip: You can deploy this as a link later (Streamlit Community Cloud).")
```
